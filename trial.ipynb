```python
from transformers import MBartForConditionalGeneration, MBartTokenizer, MBartModel
```

```python
tokenizer = MBartTokenizer.from_pretrained("facebook/mbart-large-cc25",src_lang='en_XX',tgt_lang='en_XX')
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-cc25")
```

```python
example_english_phrase = '["Q&A", "Left Right Love Destiny"], "Cecilia Cheung", "Stars", "Left Right Love Destiny"</s>[1] Who is the leading actor of the movie Left Right Love Destiny?'
expected_translation_english = "It's Cecilia Cheung"
```

```python
inputs = tokenizer(example_english_phrase, return_tensors="pt")
```

```python
inputs
```

```python
inp_tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
tokenizer.convert_tokens_to_string(inp_tokens)
```

```python
inputs['input_ids'].shape
```

```python
inputs['attention_mask'].shape
```

```python
outputs = model.generate(inputs["input_ids"],max_length=10)
```

```python
inp_tokens = tokenizer.convert_ids_to_tokens(outputs[0])
tokenizer.convert_tokens_to_string(inp_tokens)
```
